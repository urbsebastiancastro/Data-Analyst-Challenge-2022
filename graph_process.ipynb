```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_women = pd.read_csv("D:/Descargas/ng/women_in_government.csv")
print(df_women.head(10))
```

```python
df_nonsuper = pd.read_csv("D:/Descargas/ng/ratio_supervisory_and_nosupervisory_employees.csv")
print(df_nonsuper.head(5))
```

```python
# Suponiendo que ya tienes los datos cargados en un DataFrame llamado women_in_government
# Si no los tienes, asegrate de cargar los datos correctamente en un DataFrame de Pandas

# Graficar los datos
plt.figure(figsize=(10, 6))

# Trazar los datos
plt.plot(df_women['year_'], df_women['valueinthousands'], marker='o', linestyle='-')

# Aadir etiquetas y ttulo
plt.title('Number of Women in Government Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Women (in thousands)')

# Mostrar la cuadrcula
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# Suponiendo que ya tienes los datos cargados en un DataFrame llamado women_in_government
# Si no los tienes, asegrate de cargar los datos correctamente en un DataFrame de Pandas

# Filtrar solo las filas correspondientes a enero de cada ao
january_data = df_women[df_women['month_'] == 'January']

# Graficar los datos
plt.figure(figsize=(10, 6))

# Trazar los datos
plt.plot(january_data['year_'], january_data['valueinthousands'], marker='o', linestyle='-')

# Aadir etiquetas y ttulo
plt.title('Number of Women in Government Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Women (in thousands)')

# Definir los marcadores en los ejes x en incrementos de 5 aos
plt.xticks(range(min(january_data['year_']), max(january_data['year_'])+1, 5))

# Mostrar la cuadrcula
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# Suponiendo que ya tienes los datos cargados en un DataFrame llamado employee_data
# Si no los tienes, asegrate de cargar los datos correctamente en un DataFrame de Pandas

df_nonsuper_january_data = df_nonsuper[df_nonsuper['period_'] == 'M01']

# Graficar los datos
plt.figure(figsize=(10, 6))

# Trazar los datos
plt.plot(df_nonsuper_january_data['year_'], df_nonsuper_january_data['employees_ratio'], marker='o', linestyle='-')

# Aadir etiquetas y ttulo
plt.title('Number of Production or Nonsupervisory Employees and Supervisory Employees Ratio Over Time')
plt.xlabel('Year')
plt.ylabel('Employee Ratio')

# Definir los marcadores en los ejes x en incrementos de 5 aos
plt.xticks(range(min(df_nonsuper_january_data['year_']), max(df_nonsuper_january_data['year_'])+1, 5))

# Mostrar la cuadrcula
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# Suponiendo que ya tienes los datos cargados en un DataFrame llamado employee_data
# Si no los tienes, asegrate de cargar los datos correctamente en un DataFrame de Pandas

# Filtrar solo las filas correspondientes a enero de cada ao
df_nonsuper_january_data_github = df_nonsuper[df_nonsuper['period_'] == 'M01']

# Establecer el estilo de la grfica utilizando seaborn
sns.set_style("darkgrid")

# Tamao de la fuente
plt.rcParams.update({'font.size': 12})

# Graficar los datos
plt.figure(figsize=(10, 6))

# Trazar los datos
plt.plot(df_nonsuper_january_data_github['year_'], df_nonsuper_january_data_github['employees_ratio'], 
         marker='o', markersize=6, linestyle='-', linewidth=2, color='tab:blue')

# Aadir etiquetas y ttulo
plt.title('Number of Production or Nonsupervisory Employees and Supervisory Employees Ratio Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Employee Ratio', fontsize=14)

# Definir los marcadores en los ejes x en incrementos de 5 aos
plt.xticks(range(min(df_nonsuper_january_data_github['year_']), max(df_nonsuper_january_data_github['year_'])+1, 5), fontsize=12)
plt.yticks(fontsize=12)

# Mostrar la cuadrcula
plt.grid(True)

# Agregar ttulo y nota de pie de pgina
plt.figtext(.98, 0.00, 'Current Employment Statistics - CES (National)\nU.S. Bureau of Labor Statistics', ha='right', fontsize=10)

# Ajustar el diseo de la grfica
plt.tight_layout()

# Guardar la grfica como imagen
plt.savefig('employee_ratio_plot.png')

# Mostrar la grfica
plt.show()

```
